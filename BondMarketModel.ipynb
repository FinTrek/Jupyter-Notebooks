= # 1. The US corporate bond market

The [Securities Industry and Financial Markets Association (SIFMA)](https://www.sifma.org/resources/research/us-bond-market-trading-volume/) indicated that the average daily trading volume in the US corporate bond market was \\$31.9 billion during the first 10 months of 2018. The comparable figure for [equities](https://www.sifma.org/resources/research/us-equity-stats/) was \\$350 billion - ten times as high.

[Corporate bond debt outstanding](https://www.sifma.org/resources/research/us-bond-market-issuance-and-outstanding/) in second quarter 2018 was \\$9,074.4Bn according to SIFMA. The US equity total market capitalization was \\$32.121 trillion at the end of 2017 (according to the [World Bank](https://data.worldbank.org/indicator/CM.MKT.LCAP.CD?locations=US)). Thus the US stock market is 3.5 times as big as the bond market.

From these figures, we can compute the annualized turnover in the US stock market: 275%, using a 252-business-day year. The comparable annualized turnover in the US bond market was 88%.

A key difference between stock and bond markets is the continual birth-and-death cycle of bond issuance and maturity. SIFMA shows an annualized corporate bond issuance of \\$1,413.8 for the twelve months ending October, 2018, or 15.6% of end-of-period outstanding. The comparable figures for stocks were \$232.7 billion and 0.72%.

Much of the activity in the bond market is associated with new issuance. The trading firm [Market Axess](https://my.marketaxess.com/research/blog/single.php?permalink=bond-turnover-the-long-and-short-of-it-2013-09-05#.XAVPiuJReUk) found that after the initial flurry of trading, long-term turnover drops to about 25% a year: ![Market Axess](annualized-turnover-all-bonds-days-2013-09.jpg)

Thus seasoned bonds have less than one tenth of the turnover of equities, making them harder to trade. Corporate bonds mostly trade in a broker/dealer market where a small number of financial institutions - generally units of large multinational banks - engage in bilateral transactions with institutional clients. In some cases the financial institution acts as a dealer, using its own inventory and capital to facilitate trades. In other cases the financial institution is a broker, risking no capital while searching for another party to take the other side of a trade. While some electronic bond trading venues have been established, they still account for only about 1% of trading volume.

One obvious reason for the lack of modern trading infrastructure for corporate bonds is their fragmented issuance. [FINRA](http://finra-markets.morningstar.com/BondCenter) (the Financial Industry Regulatory Authority, an industry group) shows 112,300 corporate bond issues in their TRACE (Trade Reporting and Compliance Engine) in December 2018. SIFMA shows 1200 corporate bond issuers in the US, giving an average of about 94 bonds per issuer. This is a bit skewed upward as large financial corporations may have thousands of issues of short-term commercial paper that aren't intended to trade very much or at all; the original buyer almost always simply holds them to their (short) maturity.

While some companies have multiple equity issues, there is almost always a main issue of common stock that is the focus of the vast majority of that company's equity trading. Thus the likelihood that a buyer and a seller of common stock arrive at the market at the same time is very high, while the matches in the corporate bond market are attenuated. Buyers and sellers of the same issuer may arrive at the same time, but they won't necessarily be interested in the same issue.

- # 3. Market model

We assume there are three classes of investors:
- Index investors, who buy and sell only the entire market;
- Informed investors, who buy and sell individual bonds based on information; and
- Misinformed investors, who buy and sell individual bonds based on misinformation.

<u>**Simple Accounting Pricing Equation**</u> We assume that there is a market of $n$ zero-coupon credit-risky bonds. Each bond has a hazard rate $h$. Thus if $\tau$ is the time at which the bond defaults,
$$Pr_{(h,t)}(\tau\leq t)=1-e^{-ht}$$
over the horizon between now and time t from now. If a zero-coupon unit bond defaults, it pays a recovery rate $R$; typically $R\approx 40%$. Suppose there is a constant discount rate $d$. Then on a risk-neutral basis, the present value of a bond maturing at time $T$ as a function of these variables is
$$Price(h)=\bigl(R\int_0^T e^{-ds}d(1-e^{-hs})\bigr)+(1-(1-e^{-hT})))e^{-dT}=\frac{Rh}{d+h}(1-e^{-(d+h)T})+e^{-(d+h)T}$$

<u>**Beliefs and Mispricings**</u> Investors have beliefs about true price levels. If they see an item whose market price is below (above) their belief price, they buy (sell) it, up to a point. The amount they buy or sell is proportional to the size of the perceived mispricing. (At very low prices they panic and withdraw from the market.?) If $P_b$ is an investor's belief price and $P_m$ is the market price, then the purchase/sale multiplier is
$$PS(P_b,P_m,k)=min(max((P_b-P_m)k,-1),1)$$
Here $k$ is a parameter indicating how aggressively the investor acts on its beliefs.

<u>**Index Investors**</u> only buy or sell the entire market. They know (from ratings agencies) a long-term market average hazard rate, $h_{\infty}$. So (again keeping things risk-neutral), they expect market prices to average around
$$Price(h_{\infty})=\frac{R h_{\infty}}{d+h_{\infty}}(1-e^{-(d+h_{\infty})T})+e^{-(d+h_{\infty})T}$$
This isn't quite right because the average of prices is skewed. If $h_{\infty}$ is the average hazard rate, then a distribution of hazard rates around that will give a different distribution of prices. For calibration with empirical data this might be important, but for now let's just say $h_{\infty}$ is the risk-neutral average hazard rate.

Three parameters will characterize index investors' trading: $k_x$, the third argument to the PS function above; $t_x$, the transactions costs index investors pay; and $h_{\infty}$, the expected hazard rate. Of these, $t_x$ and $h_{\infty}$ can be calibrated to exogenous observations. $k_x$ is somewhat arbitrary although there is data on the volume of ETF trading.

At time 0, index investors have an endowment $N_x(0)$. The value of their account at the end of time period $t-1$ (which is the same as the beginning of time period $t$) is $N_x(t-1)$. If $P_m$ is the current market price, let $PS=PS(Price(h_{\infty}),P_m,k_x)$ &mdash; that is, drop the arguments for the rest of this paragraph. At the end of period $t-1$, index investors buy (sell if negative) an amount $N_x(t-1)\cdot PS$. It costs $t_x\cdot N_x(t-1)|PS|$ for them to do this. If $r_M(t)$ is the return on the market from end of t-1 to end of t, the update of the index account is
$$N_x(t)=N_x(t-1)\bigl(1+PS-|PS|t_x)(1+r_M(t))$$

<u>**Informed Investors**</u> purchase information in the form of a vector $Y_t=(y_{1,t},...,y_{n,t})$ giving better information on the hazard rate for each security; the hazard rate of security i is actually $h_{\infty}+y_i$. The $y_i$ follow Ornstein-Uhlenbeck random walks
$$dy_i=-\alpha y_i dt + \sigma d\beta_i$$
The $y_i$ will therefore have a tendency to return to zero if they vary too far away. Let $c_f$ be the cost of the information; this is an absolute amount.

Informed investors believe that the price of the $i^{th}$ security should be $Price(h_{\infty}+y_{i,t})$. They look at the market price $Price_i(t)$ of security i and buy or sell the proportion $PS(Price(h_{\infty}+y_{i,t}),Price_i(t),k_f)$, where $k_f$ is the aggressiveness parameter for informed investors.

At time 0, informed investors have an endowment $N_f(0)$. The value of their account at the end of time period $t$ is $N_f(t)$. At the end of time period $t-1$ (beginning of period $t$), order the individual bonds so that bond 1 is the most underpriced ($Price(h_{\infty}+y_{1,t})-Price_1(t-1)$ is largest) and bond n is the least underpriced/most overpriced.

At the end of period $t-1$, index investors buy (sell if negative) an amount $N_x(t-1)\cdot PS$. It costs $t_x\cdot N_x(t-1)|PS|$ for them to do this. Let $r_i(t)$ be the return on bond i from end of t-1 to end of t. For notational brevity let $PS_i=PS(Price(h_{\infty}+y_{i,t-1}),Price_i(t-1),k_f)$. 
The update of the informed investor account is
$$N_x(t)=N_x(t-1)\sum_{i=1}^n(1+PS_i-|PS_i|t_f)(1+r_i(t))-c_f$$



<u>**Misinformed Investors**</u> have misinformation in the form of a vector $Z_t=(z_{1,t},...,z_{n,t})$ that they think gives better information on the hazard rate for each security. The $z_i$ follow Ornstein-Uhlenbeck random walks
$$dz_i=-\alpha z_i dt + \sigma d\beta_i$$
The $z_i$ will therefore have a tendency to return to zero if they vary too far away.

Misinformed investors believe that the price of the $i^{th}$ security should be $Price(h_{\infty}+z_{i,t})$. However they are wrong about the hazard rate; the actual hazard rate is the one that informed investors know. But misinformed investors look at the market price $Price_i(t)$ of security i and buy or sell the proportion $PS(Price(h_{\infty}+z_{i,t}),Price_i(t),k_m)$, where $k_m$ is the aggressiveness parameter for misinformed investors.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

def Bond_Price(r,h,T):
    Recovery=.4
    exprh=np.exp(-(r+h)*T)
    if h==0:
        PV_Recovery=0
    else:
        PV_Recovery=h*(1-exprh)/(r+h)
    return(PV_Recovery*Recovery+exprh)
    
hrange=np.arange(0,.101,.01)
rrange=np.arange(0,.101,.01)

Z=np.zeros((len(hrange),len(rrange)))
for i,x in enumerate(hrange):
    for j,y in enumerate(rrange):
        Z[i,j]=Bond_Price(x,y,30.)
       
fig = plt.figure()
ax = fig.gca(projection='3d')

X, Y = np.meshgrid(hrange,rrange)
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(0, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
